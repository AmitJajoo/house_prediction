### Feature Enginnering

```python
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
```

```python
dataset[feature_nan].head(10)
```

```python
##replace missing value with the new label
def replace_cat_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna("Missing")
    return data
dataset = replace_cat_feature(dataset,feature_nan)
```

```python
dataset[feature_nan].isnull().sum()
```

```python
dataset[feature_nan].head(10)
```

```python
## numerical value
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes != 'O']
numerical_with_nan
```

```python
for feature in numerical_with_nan:
    print(round(dataset[feature].isnull().mean(),4)) ## Missing values percentage
```

```python
## Replacing the numerical missing value
for feature in numerical_with_nan:
    
    print(feature)
    median_value = dataset[feature].median()
    print(median_value)
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
dataset[numerical_with_nan].isnull().sum()
```

```python
dataset[numerical_with_nan].head()
```

```python
dataset.head(4)
```

```python
## Temporal variable
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] = dataset['YrSold']-data[feature]
dataset.head()
```

```python
num_feature = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
```

```python
for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
for feature in categorical_dataset:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

```python
for feature in categorical_dataset:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

### Feature Scaling

```python
feature_scaler = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scaler])
```

```python
data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),
                 pd.DataFrame(scaler.transform(dataset[feature_scaler]),columns=feature_scaler)],axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv')
```
